```python
import folium
from folium.plugins import HeatMap, HeatMapWithTime
from getdata import get_most_recent, get_day_data
```

```python
data = get_day_data('2019-02-06', scale_by_overall_max=False)
```

```python
HeatMapWithTime(data[0], index=data[1], radius=30).add_to(campus)
```
